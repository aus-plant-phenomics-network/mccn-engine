```python
import matplotlib.pyplot as plt

from mccn.client import MCCN
from mccn.extent import GeoBoxBuilder

ENDPOINT = "http://203.101.227.130:8082/"
collection = "OZBarleyTestData"
```

### Full dataset

```python
client = MCCN(ENDPOINT, collection, shape=100)
```

```python
ds = client.load()
```
